Temos em mãos um arquivo com dados de pacientes que desenvolveram ou não diabetes. Então devemos gerar um novo arquivo CSV e encaminhar para o Cientista de Dados.

```python
%%sql

CREATE TABLE pacientes (Pregnancies INT,
                        Glucose INT,
                        BloodPressure INT,
                        SkinThickness INT,
                        Insulin INT,
                        BMI DECIMAL(8, 2),
                        DiabetesPedigreeFunction DECIMAL(8, 2),
                        Age INT,
                        Outcome INT);
```

```python
%%sql

SELECT * FROM pacientes
```

```python
%%sql

INSERT INTO pacientes(Pregnancies, 
                      Glucose, 
                      BloodPressure, 
                      SkinThickness, 
                      Insulin, 
                      BMI, 
                      DiabetesPedigreeFunction, 
                      Age, 
                      Outcome) 
SELECT Pregnancies, 
       Glucose, 
       BloodPressure, 
       SkinThickness, 
       Insulin, 
       BMI, 
       DiabetesPedigreeFunction, 
       Age, 
       Outcome 
FROM diabetes WHERE Age > 50;
```

```python
%%sql

SELECT * FROM pacientes
```

```python
%%sql 

ALTER TABLE pacientes
ADD Perfil VARCHAR(10);
```

```python
%%sql

SELECT * FROM pacientes
```

```python
%%sql

UPDATE pacientes
SET Perfil = 'Normal'
WHERE BMI < 30;
```

```python
%%sql

SELECT * FROM pacientes
```

```python
%%sql

UPDATE pacientes
SET Perfil = 'Obeso'
WHERE BMI >= 30;
```

```python
%%sql

SELECT * FROM pacientes
```

## Carregando os Dados no Pandas e Salvando o CSV

```python
# Query
query = cnn.execute("SELECT * FROM pacientes")
query
```

```python
# List Comprehension
cols = [coluna[0] for coluna in query.description]
cols
```

```python
# Gera o dataframe
resultado = pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```

```python
# Shape
resultado.shape
```

```python
# Visualiza
resultado.head()
```

```python
# Salva em CSV
resultado.to_csv('dataset/pacientes.csv', index = False)
```
